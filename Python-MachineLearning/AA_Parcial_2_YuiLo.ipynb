```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
import seaborn as sns
# Importar dataset
customer_data = pd.read_csv(path+'/data/clientesDatos.csv')
customer_data.drop(['ID'], axis = 'columns', inplace=True)
customer_data
```

```python
# crea un pairplot, una matriz con varias gráficas, para observar como se
#relacionan las dimensiones del dataset cuando se grafican en pares. Se utiliza la librería Seaborn.
sns.set(font_scale=1.1)
sns.set_style('whitegrid')
grid = sns.pairplot(data=customer_data, vars=customer_data.columns[1:8], hue='Occupation')
```

```python
#Reducción de las Dimensiones mediante Análisis de Componentes Principales (PCA)
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=11)
pca.fit(customer_data)
customer_data.shape
```

```python
#Transformado el Dataset clentesdatos de 8 dimensiones a 2 dimensiones
customer_pca = pca.transform(customer_data)
customer_pca.shape
```

```python
#Creando un Estimador KMeans
#Entrenamiento del Modelo
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def kmeans_grafica(k):
  kmeans=KMeans(n_clusters=k, random_state=11)
  kmeans.fit(customer_data)
  #Estudiando la distribución original de clases en el dataset
  '''
  print("--------------Agrupación para k = "+str(k))
  print(kmeans.labels_[0:500])
  print(kmeans.labels_[500:1000])
  print(kmeans.labels_[1000:1500])
  print(kmeans.labels_[1500:2000])'''
  #Visualización de los datos con su nueva representación.
  customer_pca_df = pd.DataFrame(customer_pca, columns=['component1', 'component2'])
  customer_pca_df['Occupation'] = kmeans.labels_
  axes = sns.scatterplot(data=customer_pca_df, x='component1', 
                         y='component2', hue='Occupation', legend='brief', palette="deep") 
  customer_centers = pca.transform(kmeans.cluster_centers_)
  dots = plt.scatter(customer_centers[:,0], customer_centers[:,1], s=100, c='Red')
```
